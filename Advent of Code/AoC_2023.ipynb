```python
def preprocessing_d4(input:str):
    input = input.replace('  ',' ')
    if input.startswith(' '):
        input=input[1:]
    if input.endswith(' '):
        input=input[:-1]
    return input

def d4_p1(input:str,verbose:bool=False)->int:
    ans=0
    card_num=1
    for card in input.split('\n'):
        winning, options = card.split(':')[1].split('|')
        winning_set = set(preprocessing_d4(winning).split(' '))
        options_set = set(preprocessing_d4(options).split(' '))
        match = len(winning_set & options_set)
        if verbose:
            # print(f'Card: {card_num}',len(winning_set),len(options_set),match)
            print(f'Card: {card_num}',match, winning_set & options_set)
            # if len(options_set)==26:
            #     print(f'Card: {card_num}',len(winning_set),options_set)
            # card_num+=1
        if match>0:
            ans+=2**(match-1)
    print(f'Answer is: {ans}')
```

```python
def update_card_nums_stack(card_nums_stack:list, cur_card_num:int, match:int)->list:
    for i in range(min(len(card_nums_stack),match)):
        card_nums_stack[i]+=cur_card_num
    if len(card_nums_stack)<match:
        card_nums_stack = card_nums_stack+[cur_card_num]*(match-len(card_nums_stack))
    return card_nums_stack

def d4_p2(input:str,verbose:bool=False)->int:
    ans=0
    card_nums=1
    card_nums_stack=[0]
    for card in input.split('\n'):
        winning, options = card.split(':')[1].split('|')
        winning_set = set(preprocessing_d4(winning).split(' '))
        options_set = set(preprocessing_d4(options).split(' '))
        match_num = len(winning_set & options_set)
        cur_card_num = 1 + card_nums_stack.pop(0) if card_nums_stack else 1
        # add new copys to the stack
        card_nums_stack=update_card_nums_stack(card_nums_stack,cur_card_num,match_num)
        
        if verbose:
            print(f'Card num: {card_nums}')
            print(cur_card_num, match_num)
            print(card_nums_stack)

        # update ans 
        ans+=cur_card_num
        card_nums+=1
    print(f'Answer is: {ans}')
```

The number of next card depends on the number of current card and the matching number.  
We can use a stack to save the number of the winning copy cards and update it like:
Assuming we have X *Current card* and the matching number is Y,  
we will update the stack in a way that the first Y elements increase X:  
`[a,b,c,d,...,x,y,z] -> [a+X, b+X,...,f+X,g,...,x,y,z]`(if the stack has more than Y elements) OR  
`[a,b,c] -> [a+X, b+X, c+X, X,...,X]`(Adding new elements with value X and making it have Y elements if the stack has less than that)

```python
# Construct converted mapping structure in the following format:
# [[source_start, source_end, destination_start]]
def create_converted_mapping(mapping:list[str])->list[list[int]]:
    converted_mapping=[]
    for row in mapping:
        destination_start, source_start, length = row.split(' ')
        # print(destination_start, source_start, length)
        converted_mapping.append([int(source_start), int(source_start)+int(length)-1, int(destination_start)])
    return converted_mapping

def d5_get_mappings(input:str)->list[list[int]]:

    # Get seeds 
    seeds = re.match(r'seeds: ([\d ]*)',input)[1].split(' ')
    output = [seeds]
    
    # Get all maps
    seed2soil = re.search(r'seed-to-soil map:\n([^a-z]*)',input)[1][:-2].split('\n')
    output.append(create_converted_mapping(seed2soil))
    soil2fertilizer = re.search(r'soil-to-fertilizer map:\n([^a-z]*)',input)[1][:-2].split('\n')
    output.append(create_converted_mapping(soil2fertilizer))
    fertilizer2water= re.search(r'fertilizer-to-water map:\n([^a-z]*)',input)[1][:-2].split('\n')
    output.append(create_converted_mapping(fertilizer2water))
    water2light= re.search(r'water-to-light map:\n([^a-z]*)',input)[1][:-2].split('\n')
    output.append(create_converted_mapping(water2light))
    light2temperature= re.search(r'light-to-temperature map:\n([^a-z]*)',input)[1][:-2].split('\n')
    output.append(create_converted_mapping(light2temperature))
    temperature2humidity= re.search(r'temperature-to-humidity map:\n([^a-z]*)',input)[1][:-2].split('\n')
    output.append(create_converted_mapping(temperature2humidity))
    humidity2location= re.search(r'humidity-to-location map:\n([^a-z]*)',input)[1].split('\n')
    output.append(create_converted_mapping(humidity2location))
    return output

def source_to_dest(source:list[int],mapping:list[list[int]],verbose:bool=False)->list[int]:
    if verbose:
        print(f'mapping: {mapping}')
    next_map=[]
    for s in source:
        if verbose:
            print(f'source: {s}')
        find_map=False
        for row in mapping:
            if row[0]<=s<=row[1]:
                # print("in between")
                next_map.append(s-row[0]+row[2])
                find_map=True
                break
        if not find_map:
            next_map.append(s)
        if verbose:
            print(f'map to: {next_map[-1]}')
    if verbose:
        print(next_map)
    return next_map
```

```python
def get_d5_p2_seeds(seeds:list[str])->list[int]:
    output=[]
    for i in range(0,len(seeds),2):
        start,end =int(seeds[i]),int(seeds[i+1])
        output.append((start,start+end-1))
    ############# deprecated
    # for i in range(0,len(seeds),2):
    #     start,end =int(seeds[i]),int(seeds[i+1])
    #     # print(start,end)
    #     output+=[start+j for j in range(end)]
    ##############
    return output

def source_to_dest_ranges(source:list[tuple[int]],mapping:list[list[int]],verbose:bool=False)->list[tuple[int]]:
    output=[]
    while source:
        start,end = source.pop()
        if verbose:
            print(f'source : [{start}, {end}]')
        for source_start, source_end, destination_start in mapping:
            if verbose:
                print('current mapping group: ', source_start, source_end, destination_start)
            find_map=False
            if start>=source_start and start<=source_end:
                dest_start = start-source_start+destination_start
                # [start,end] within one group
                if end<=source_end:                
                    dest_end = end-start+dest_start
                    output.append((dest_start, dest_end))
                    if verbose:
                        print(f'within one group, from [{start},{end}] to [{dest_start},{dest_end}]')
                else:
                # [start,group_end]   [group_end+1, end] into two groups
                    group_end = source_end
                    dest_end = source_end-source_start+destination_start
                    source.append((group_end+1,end))
                    output.append((dest_start, dest_end))
                    if verbose:
                        print(f'within two group, the first part is from [{start},{group_end}] to [{dest_start},{dest_end}]')
                        print(f'the second part(match needed) is [{group_end+1}, {end}]')
                find_map=True
                break
        if not find_map:
            # not in any group
            output.append((start,end))
            if verbose:
                print(f'not in any group, value unchanged, keeping as [{start},{end}]')
    if verbose:
        print(f'mapping to: {output}')
    return output

def d5_p2(input:str,verbose:bool=False)->int:
    mappings = d5_get_mappings(input)
    # get seeds
    seeds = get_d5_p2_seeds(mappings[0])
    if verbose:    
        print(f'seeds: {seeds}')
    for mapping in mappings[1:]:
        seeds = source_to_dest_ranges(seeds, mapping, verbose)
    if verbose:
        print(f'final locations: {seeds}')
    ans = min([i[0] for i in seeds])
    print(f'final ans: {ans}')
    return ans
```

```python
def push_rock(input:str,reverse:bool=False)->str:
    row,col = row,col = len(input.split('\n')), len(input.split('\n')[0])
    next_one = []
    for i in range(col):
        whole_col = input[i::col+1]
        if reverse:
            whole_col=whole_col[::-1]
        # print(whole_col)
        # load=row
        next_col=''
        s_num=d_num=0
        for j in whole_col:
            if j=='O':
                s_num+=1
            elif j=='.':
                d_num+=1
            else:
                next_col+='O'*s_num+'.'*d_num+'#'
                s_num=d_num=0
        next_col+='O'*s_num+'.'*d_num
        # print(next_col)
        if reverse:
            next_col = next_col[::-1]
        next_one.append(next_col)
    return ('\n').join(next_one)

def push_rock_cycle(input:str)->str:
    # print(input)
    # print('-----to north-------')
    n_input = push_rock(input)
    # print(n_input)
    # print('-----to west----')
    n_input = push_rock(n_input)
    # print(n_input)
    # print('-----to south----')
    n_input = push_rock(n_input,True)
    # print(n_input)
    # print('-----to east----')
    n_input = push_rock(n_input,True)
    # print(n_input)
    return n_input
```
